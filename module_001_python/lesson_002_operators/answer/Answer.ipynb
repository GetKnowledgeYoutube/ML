Фильтр по квартирам. Есть информация о квартире, проверить следующие условия:
* Цена меньше 3млн и 2 комнатная, но не в центре
* В центре, не на первых этажах (1-3) и должен быть лифт
* Без лифта и цена от 1млн до 2.5 млн

```python
# Цена квартиры
price = 2_000_000

# Кол-во комнат
rooms = 1

# Расположение в городе
location = "center"

# Наличие лифта
elevator = True

# Этаж
floor = 7
```

```python
# Пишите здесь свой код

expr1 = price < 3_000_000 and rooms == 2 and location != "center"
expr2 = location == "center" and floor > 3 and elevator
expr3 = not elevator and price >= 1_000_000 and price <= 2_500_000

print(expr1)
print(expr2)
print(expr3)
```
