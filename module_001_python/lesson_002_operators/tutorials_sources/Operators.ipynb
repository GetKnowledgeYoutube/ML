# 1 - if, else, elif

Основные логические операторы ветвления

```python
month = 15

if month <= 0 or month > 12:
    print('Такого месяца нет')
else:
    if month == 1 or month == 2 or month == 12:
        print('Зима наступила')
    elif month >= 3 and month < 6:
        print('Скучаем по зиме')
    elif month >=6 and month < 9:
        print('Когда же уже зима')
    else:
        print('Зима близко')
```

Основные логические операции:
* Операции сравнения
* and - истина когда оба операнда истина
* or - истина когда хотябы один операнд истина
* not - превращает истину в ложь и наоборот

```python
score = 0.73
age = 21
conviction = True # судимость
official = True # чиновник

if official or (score > 0.7 and age >= 18 and not conviction):
    print('Выдать кредит')
else:
    print('Не выдавать кредит')
```

Типы данных могут автоматически преобразовываться в логические выражения

# 2 - Loops

```python
for i in range(1,10):
    print(i)
```

```python
for i in range(1, 10, 3):
    print(i)
```

```python
for i in range(1,11):
    if i % 2:
        continue
    print(i)
```

```python
for i in range(1,11):
    if i == 8:
        break
    print(i)
```

```python
message = "Hello"
for c in message:
    print(c)
```

```python
i = 0
while i < 10:
    print(i)
    i += 1
```

```python
i = 0
while True:
    if i == 30:
        break
    print(i)
```
